#몬테카를로
몬테카를로 방법을 이용하여 $f(x) = x^2$ 함수의 넓이비율을 알아보도록 합시다.

$\\$

--------------------------

$\\$

```
import matplotlib.pyplot as plt
import random
import math
import numpy as np
```
필요한 모든 모듈을 불러옵니다.   

```
ranx = []
rany = []
t=1000
inf = []
```

ranx 는 무작위로 불러온 x값들의 리스트입니다.  
마찬가지로 rany은 무작위로 불러온 y값들의 리스트입니다.  
t는 무작위로 불러온 x와 y의 값들의 갯수를 말합니다. 여기서는 1000개의 x값과 y값들을 무작위로 불러옵니다.  
inf는 함수값보다 작은 x값들의 리스트입니다.

$\\$

```
for n in range(0,t):
  x_ran = random.random()
  ranx.append(x_ran)
  
  y_ran = random.random()
  rany.append(y_ran)
  
  if y_ran <= x_ran**2:
    inf.append(y_ran)
 ```
t만큼의 x값과 y값들을 불러오기 위해 `range`를 0부터 t미만으로 설정합니다.  
`random.random()`함수로 x값과 y값들을 무작위로 불러온 후에 각가 ranx와 rany 리스트에 넣습니다.  
만일 무작위로 찍은 점의 y값이 함수값보다 작을 경우에는 inf 리스트에 들어가도록 합니다.

$\\$

```
x=np.linspace(0,1)
plt.plot(x,x**2)
plt.scatter(ranx,rany)
```
정의역을 0과 1 사이로 정의합니다. 치역은 x값의 제곱으로 정의하여 그래프를 그립니다.  
x값이 ranx, y값이 rany인 점들을 산점도로 나타냅니다.

$\\$

```
prop = len(inf)/t
prop
```
prop은 함수값보다 같거나 작은 점들과 전체 점들의 비율을 계산합니다. 

````python
import matplotlib.pyplot as plt
import random
import math
import numpy as np

ranx = []
rany = []
t=1000
inf = []

for n in range(0,t):
  x_ran = random.random()
  ranx.append(x_ran)
  
  y_ran = random.random()
  
  if y_ran <= x_ran**2:
    inf.append(y_ran)

  rany.append(y_ran)

x=np.linspace(0,1)
plt.plot(x,x**2)
plt.scatter(ranx,rany)

prop = len(inf)/t
prop
````
